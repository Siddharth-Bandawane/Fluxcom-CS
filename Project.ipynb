```python
import pandas as pd
import numpy as np
```

```python
tall_twr= pd.read_csv("AMF_US-PFa_BASE_HR_26-5.csv", skiprows=2)
tall_twr.columns
```

```python
# Convert timestamp strings to integer format
start_timestamp = int("201908010000")
end_timestamp = int("201908012300")

# Filter data based on integer timestamps
tall_twr = tall_twr[(tall_twr.TIMESTAMP_START >= start_timestamp) & (tall_twr.TIMESTAMP_START <= end_timestamp)]
```

```python
# Define the columns you want to keep
columns_to_keep = ['TIMESTAMP_START', 'TA_1_1_1', 'TA_1_2_1', 'TA_PI_F_1_3_1','WS_1_1_1', 'WS_1_2_1', 'WS_PI_F_1_3_1']

# Select only the desired columns
ds = tall_twr[columns_to_keep]
ds = ds.reset_index(drop=True)
data=ds.values
```

```python
Cp = 1003.62  # Specific heat
k = 0.41  # von Karman constant
g = 9.81  # Gravitational acceleration
P_atm = 1013.25  # hPa
alpha = 0.286  # Exponent for potential temperature calculation
alr = g / Cp  # Adiabatic lapse rate
z_ref = 396  # Reference height
```

```python
# Define velocity and temperature data
u = data[:,4:7]
T = data[:,1:4] + 273.15

# Measurement heights
z = np.array([30,122,396])
```

```python
# Convert all Temps to potential temperature
pT = np.zeros_like(T)
for w in range(len(z)):
    pT[:, w] = T[:, w] + g / Cp * z[w]
#pT = (data(:,1:4)+273.15)*(1000/P_atm)^alpha; %Convert to K and potential temp. (Alternative formulation)
#Data for PA available
```

```python
# Calculate geometric mean height
zm = np.zeros(len(z) - 1)
for i in range(len(z) - 1):
    zm[i] = np.sqrt(z[i] * z[i + 1])
```

```python
# Logarithmic approximation for vertical gradient of velocity and Potential Temperature and calculation of Richardson Number
gradU = np.zeros((len(u), len(u[0]) - 1))
gradT = np.zeros((len(T), len(T[0]) - 1))
Ri = np.zeros_like(gradU)
for i in range(len(u[0]) - 1):
    gradU[:, i] = (u[:, i + 1] - u[:, i]) / (zm[i] * np.log(z[i + 1] / z[i]))
    gradT[:, i] = (T[:, i + 1] - T[:, i]) / (zm[i] * np.log(z[i + 1] / z[i])) + g / Cp
    Ri[:, i] = (g / T[:, i]) * gradT[:, i] / (gradU[:, i]) ** 2
```

```python
# Sort into stable and unstable categories: use only profiles that yield 2 positive or 2 negative Ri values
unstable = []
stable = []
for i in range(len(data)):
    if Ri[i, 0] < 0 and Ri[i, 1] < 0:
        unstable.append([i, *Ri[i]])
    elif 0 <= Ri[i, 0] and 0 <= Ri[i, 1] :
        stable.append([i, *Ri[i]])
```
